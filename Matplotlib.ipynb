# Installation et importation du package

```python
#Importation des packages
import matplotlib.pyplot as plt
import numpy as np
```

# Tracé de graphiques simples avec Matplotlib

```python
# Génération de données
x=np.linspace(0,10,1000)
y=np.exp(x)
```

```python
# Tracer un graphique simple
plt.plot(x,y)
plt.show()
```


# Personnalisation de graphiques avec Matplotlib

```python
#Changer la taille d'un graphique avec la largeur et la hauteur de la figure en pouce
plt.figure()
plt.plot(x,y)
```

```python
# Titre et axes
plt.plot(x,y)
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.show()
```

```python
# Changement de Couleur
plt.plot(x,y,c="red")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.show()
```

```python
# Changer le type de ligne
plt.plot(x,y,c="red",ls="--")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.show()
```

```python
# Ajouter une Légende

plt.plot(x,y,c="red",ls="-",label="exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.show()
```

```python
# Ajouter une grille sur le graphique
plt.plot(x,y,c="red",ls="-",label="exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.grid()
plt.show()
```

## Faire des graphique sur un même carré

```python
# Créer un graphique vide
z=2*y
y3=z*2
plt.plot(x,y,c="red",ls="-",label="exp")
plt.plot(x,z,c="blue",ls="-",label="2*exp")
plt.plot(x,y3,c="green",ls="-",label="4*exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.show()
```

```python
# Changer les limites des axes
z=2*y
y3=z*2
plt.plot(x,y,c="red",ls="-",label="exp")
plt.plot(x,z,c="blue",ls="-",label="2*exp")
plt.plot(x,y3,c="green",ls="-",label="4*exp")
plt.xlim((6,10))
plt.ylim((1000,80000))
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.show()
```

# Marche aléatoire

```python
# Marche aléatoire avec Matplotlib
# Importation des packages nécessaires  
# !pip install matplotlib
# NumPy  import, seed
import numpy as np
np.random.seed(1230)  # Pour la reproductibilité des résultats
# Marche aléatoire      
 
random_walk = [0]  # Initialization pour la marche aléatoire

for x in range(100) :        # 100 étapes de la marche aléatoire
    step = random_walk[-1] # Dernière position de la marche aléatoire
    # Lancer un dé à 6 faces
    dice = np.random.randint(1,7)

    if dice <= 2:
        step = max(0, step - 1) # Ne pas descendre en dessous de 0
    elif dice <= 5:
        step = step + 1   # Avancer d'une case
    else:
        step = step + np.random.randint(1,7)  # Avancer de 1 à 6 cases

    random_walk.append(step)  # Ajouter la nouvelle position à la marche aléatoire

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```

```python
# Initialize all_walks (don't change this line)
all_walks = []

# Simulate random walk five times
for i in range(5) :

    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk) 

# Print all_walks
print(all_walks)
```

```python
# initialize and populate all_walks
all_walks = []
# Simuler une marche aléatoire cinq fois
for i in range(5) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)

print(all_walks)
```

```python
# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()
# Clear the figure
#plt.clf()

```

```python
# Il y a une "erreur" dans le code précédent, il faut transposer np_aw avant de le tracer
# Car nous voulons que chaque ligne représente une marche aléatoire différente
# Sinon, chaque colonne représenterait une marche aléatoire différente

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

```python
# Simulater une marche aléatoire 20 fois
all_walks = []
for i in range(20) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implementer une probabilité de 0.5% de réinitialiser la marche aléatoire
        # Si la probabilité est atteinte, réinitialiser step à 0
        if np.random.rand() <= 0.005 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

```python
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```

```python
# Calculer la probabilité d'atteindre au moins 60 marches de haut
prob_60 = np.sum(ends >= 60) / len(ends)
print(f"Probabilité d'atteindre au moins 60 marches : {prob_60:.2f}")
```
